# Exhaustive Search Method

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x = np.linspace(6.5,7.5,11)
y = np.linspace(8.5,9.5,11)
xv, yv = np.meshgrid(x, y)
```

```python
def objective(x,y): 
    return (x-7)**2 + (y-9)**2
```

```python
V = objective(xv,yv)
# V
```

```python
fig = plt.figure(figsize=(10,8),dpi=100)
ax = sns.heatmap(V, annot=True,xticklabels=x,yticklabels=y)
plt.xlabel("x1")
plt.ylabel("x2")
plt.show()
```
